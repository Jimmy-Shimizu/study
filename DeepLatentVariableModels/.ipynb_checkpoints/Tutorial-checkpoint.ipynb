```
class MixtureOfMultinomials:
    def __init__(self, config):
        self.config = config

    def model(self, data):
        V, C = self.config
        N, T = data.shape
        mu = pyro.param("mu", torch.rand(C), 
                        constraint=constraints.simplex) 
        pi = pyro.param("pi", torch.rand(C, V),
                        constraint=constraints.simplex)
        
        with pyro.iarange("n", N, subsample_size=BATCH) as ind:
            z = pyro.sample("z", ds.Categorical(mu.expand(BATCH, C)), 
                           infer=dict(enumerate="parallel"))

            for t in pyro.irange('T', T):
                x= pyro.sample(f"x_{t}", ds.Categorical(pi[z]),
                               obs=data[ind, t])
           
    
```

```
# @title MultiRNN Implementation
# Dimension in Pyro are [N, C, V]
class MultiRNN(nn.Module):
    "Compute multiple different RNNs in batch."
    def __init__(self, C, V, hidden_dim):
        super(MultiRNN, self).__init__()
        self.E = nn.Embedding(V, hidden_dim)
        self.V = nn.Linear(hidden_dim, V)
        self.W = nn.Parameter(torch.zeros(C, hidden_dim, hidden_dim))
        self.U = nn.Parameter(torch.zeros(C, hidden_dim, hidden_dim))
        self.hidden_dim = hidden_dim
 
    def forward(self, x, z, h):
        if h is None:
            h_0 = torch.zeros(z.shape[0], 1, self.hidden_dim)
            return self.V(h_0), h_0
        e = self.E(x)
        h_t = torch.tanh(torch.einsum("nckh,ch->nck", [self.W[z], e]) 
                         + torch.einsum("nckh,nch->nck",[self.U[z],h]))
        return self.V(h_t), h_t              
```

```

class MixtureOfRNNs:
    def __init__(self, config):
        self.config = config
        V, C = config
        self.rnn = MultiRNN(C, V, 10)
        
    def model(self, data):
        V, C = self.config
        N, T = data.shape
        rnn = pyro.module("rnn", self.rnn)
        mu = pyro.param("mu", torch.rand(C),
                        constraint=constraints.simplex) 
        with pyro.iarange("n", N, subsample_size=BATCH) as ind:
            z = pyro.sample("z", ds.Categorical(mu.expand(BATCH, C)),
                            infer=dict(enumerate="sequential"))
            h = None
            for t in range(T):                
                out, h = rnn(data[ind, t-1], z, h)
                pyro.sample(f"x_{t}", 
                            ds.Categorical(logits=out),
                            obs=data[ind, t])

           
```

```
class TopicVector:
    def __init__(self, config):
        self.config = config
        V, C = config
        
    def model(self, data):
        V, C = self.config
        N, T = data.shape
        alpha = pyro.param("alpha", torch.rand(N, C),
                           constraint=constraints.positive) 
        with pyro.iarange("n", N, subsample_size=BATCH) as ind:
            z = pyro.sample("z", ds.Dirichlet(alpha).independent(1))
            h = None
            for t in range(T):
                pyro.sample(f"x_{t}", 
                            ds.Categorical(logits=z),
                            obs=data[ind, t])
```

```
#@title Simple RNN Implementation
class SimpleRealRNN(nn.Module):
    def __init__(self, C, V, hidden_dim):
        super(SimpleRealRNN, self).__init__()
        self.E = nn.Embedding(V, hidden_dim)
        self.V = nn.Linear(hidden_dim, V)
        self.hidden_dim = hidden_dim
        self.W = nn.Linear(hidden_dim, hidden_dim)
        self.U = nn.Linear(hidden_dim, hidden_dim)
        self.Q = nn.Linear(C, hidden_dim)
 
    def forward(self, x, z, h):
        if h is None:
            h_0 = torch.zeros(1, self.hidden_dim)
            return self.V(h_0), h_0
        e = self.E(x)
        h_t = torch.tanh(self.W(e) 
                         + self.U(h)
                         + self.Q(z))
        return self.V(h_t), h_t        
 
```

```

class NeuralTopicVector:
    def __init__(self, config):
        self.config = config
        V, C = config
        self.rnn = SimpleRealRNN(C, V, 10)
        
    def model(self, data):
        V, C = self.config
        N, T = data.shape
        rnn = pyro.module("rnn", self.rnn)
        mu = pyro.param("mu", torch.rand(N, C)) 
        Sigma = pyro.param("Sigma", torch.rand(N, C),
                          constraint=constraints.positive) 
        with pyro.iarange("n", N, subsample_size=BATCH) as ind:
            z = pyro.sample("z", ds.Normal(mu[ind], Sigma[ind]).independent(1))
            h = None
            for t in range(T):
                out, h = rnn(data[ind, t-1], z, h)
                pyro.sample(f"x_{t}", 
                            ds.Categorical(logits=out),
                            obs=data[ind, t])
        
```

```
class HiddenMarkovModel:
    def __init__(self, config):
        self.config = config
        
    def model(self, data):
        V, C = self.config
        N, T = data.shape
        mu_0 = pyro.param("mu_0", torch.rand(C),
                       constraint=constraints.simplex) 
        mu = pyro.param("mu", torch.rand(C, C),
                       constraint=constraints.simplex) 
        pi = pyro.param("pi", torch.rand(C, V), 
                       constraint=constraints.simplex)
        with pyro.iarange("n", N, subsample_size=BATCH) as ind:
            for t in range(T): 
                param = mu[z_last] if t > 0 else mu_0.expand(BATCH, C)
                z = pyro.sample(f"z_{t}", ds.Categorical(param),
                               infer=dict(enumerate="sequential"))
             
                pyro.sample(f"x_{t}", ds.Categorical(pi[z]),
                            obs=data[ind, t])
                
                z_last = z        
```

## Variational Inference

### Amortized q 

```
#@title Amortized
pgm = daft.PGM([3, 3])
pgm.add_node(daft.Node("z", "$z^{(n)}$", 1, 1, observed=False))
pgm.add_node(daft.Node("lambda", r"$\lambda$", 1, 2, fixed=True))
pgm.add_node(daft.Node("x", "$x^{(n)}$", 2, 1, observed=True))

pgm.add_edge("lambda", "z")
pgm.add_edge("x", "z")

pgm.add_plate(daft.Plate([0.5, 0.5, 2, 1], ""))
pgm.render();pgm.figure.dpi=150
```

```
class Encoder(nn.Module):
    "With parameters \lambda"
    def __init__(self, C, V, hidden_dim):
        super(Encoder, self).__init__()
        self.E = nn.Embedding(V, hidden_dim)
        self.lstm = nn.LSTM(hidden_dim, hidden_dim,
                            batch_first=True)
        self.M = nn.Linear(hidden_dim, C)
        self.S = nn.Linear(hidden_dim, C)
        
        
    def forward(self, data):
        out = self.lstm(self.E(data))[1][0]
        return self.M(out), self.S(out).exp()
```

```

class AmortizedNormal:
    def __init__(self, config):
        self.config = config
        V, C = config
        self.encoder = Encoder(C, V, 50)
        
    def guide(self, data):
        V, C = self.config
        N, T = data.shape
        encoder = pyro.module("encoder", self.encoder)
       
        with pyro.iarange("n", N, subsample_size=BATCH) as ind:
            mu, Sigma = encoder(data[ind])
            z = pyro.sample("z", ds.Normal(mu, Sigma).independent(1))
```

```
class AmortizedDiscrete:
    def __init__(self, config):
        self.config = config
        V, C = config
        self.encoder = Encoder(C, V, 50)
        
    def guide(self, data):
        V, C = self.config
        N, T = data.shape
        encoder = pyro.module("encoder", self.encoder)
        with pyro.iarange("n", N, subsample_size=BATCH) as ind:
            mu, _ = encoder(data[ind])
            z = pyro.sample("z", ds.Categorical(logits=mu))
```

```

class MeanFieldDiagonalNormal:
    def __init__(self, config):
        self.config = config
        
    def guide(self, data):
        V, C = self.config
        N, T = data.shape
        lambd_mu = pyro.param("lambda_mu", torch.rand(N, C)) 
        lambd_sigma = pyro.param("lambda_sigma", torch.rand(N, C),
                                     constraint=constraints.positive)

        with pyro.iarange("n", N, subsample_size=BATCH) as ind:
            z = pyro.sample("z", ds.Normal(lambd_mu[ind],
                                           lambd_sigma[ind]).independent(1))
```

## Inference

```
import pyro.infer
import pyro.optim
from torch.distributions import constraints
```

```
def optimize(model, guide, trace, data):
    pyro.clear_param_store()
    optimizer = pyro.optim.Adam({"lr": 1.0e-3})
    svi = pyro.infer.SVI(model, guide, optimizer, 
              loss=trace)
    losses = []
    for epoch in range(1000):
        losses.append(svi.step(train_iter.text.transpose(0, 1)) / 
                      (train_iter.text.shape[1] * train_iter.text.shape[0]))
    return losses
```
