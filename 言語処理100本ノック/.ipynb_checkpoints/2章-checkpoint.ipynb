12. 1列目をcol1.txtに，2列目をcol2.txtに保存

```python
first = []
second = []
for line in l:
    row = line.split("\t")
    first.append(row[0])
    second.append(row[1])
str1 = '\n'.join(first)
str2 = '\n'.join(second)
with open('data/col1.txt', 'w') as f:
    f.write(str1)
with open('data/col2.txt', 'w') as f:
    f.write(str2)
```

```python
!cut -f 1 data/popular-names.txt
```

13. col1.txtとcol2.txtをマージ

```python
with open('data/col1.txt', 'r') as f:
    l_1 = f.read()
with open('data/col2.txt', 'r') as f:
    l_2 = f.read()
first = l_1.split()
second = l_2.split()
with open('data/col1_2.txt', 'w') as f:
    a = ""
    for i in range(len(first)):
        a += first[i] + "\t" + second[i] + "\n"
    f.write(a)
```
