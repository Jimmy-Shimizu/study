#### モデルベース
- 線形次元削減や非縁行列因子分解と本質的には同じ構造  
- 似た傾向を持つユーザーのデータを元に欠損値を補完  
- 潜在的なアイテムの特徴とユーザーの嗜好を抽出できる
    - 個々のアイテム・ユーザーを参照せず全体の購買履歴やレーティングのみから推薦が行える

時系列データを利用することで任意の時刻における特徴dの流行度を考慮できる

### ベイズ学習のこれから
##### ベイズ学習とは
- ベイズ学習は不確実性を伴うシステムの挙動解析
- 条件付け、周辺化によって知りたい事象の分布を理解する試み  

#### ベイズ学習の今後
- 大規模かつ非線型性の高いモデルの効率学習が求められている
- ベイズ学習は
    - 適用範囲の広さ
    - 大量データに対する複雑なモデルの効率計算  
    
深層学習と組み合わさることによりさらなる応用領域を拡大
